# DAC Noise Simulation

```python
# Library Imports and Python parameter settings
%matplotlib inline
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

debugme = 1
# Update the matplotlib configuration parameters:
plt.rcParams.update({'font.size': 20,
                     'font.family': 'serif',
                     'figure.figsize': (10, 8),
                     'axes.grid': True,
                     'grid.color': '#555555'})
```

```python
# Make the signal which goes to the DAC
fs = 2048     # sample rate
f0 = 9.731   # sine wave frequency
t_final = 64   # duration of signal
A = .31e4      # signal amplitude in DAC counts

t = np.linspace(1/fs, t_final, t_final*fs)
y0 = A * (np.sin(2*np.pi*f0*t) + 0.0*np.sin(2*np.pi*(1.43*f0)*t))

ndither = np.random.normal(0, 0.5, np.size(t))
y1 = y0 + ndither

# Quantize the sine waves
q0 = np.floor(y0)
q1 = np.floor(y1)

# make the difference
n0 = y0 - q0
n1 = y1 - q1


N = 0.0021*t_final*fs
```

```python
fig = plt.figure(figsize=(16, 9))
#plt.loglog(aligo[:,0], sqrt(aligo[:,1]), color='Indigo', ls='--', alpha=0.65, lw=4)

plt.subplot(211)
plt.plot(t[0:N], y0[0:N], linestyle='-', lw=4, alpha=0.97, color='Orange',
             label='Signal')
plt.plot(t[0:N], q0[0:N], linestyle='-', lw=2, alpha=0.47, color='Blue',
             label='Quantized Signal')

#plt.xlabel(r'Time [s]', fontsize=18)
#plt.ylabel(r'Strain', fontsize=18)

#plt.title("IR Transmission during ALS Scan");
plt.grid(b=True, which="major", color='gray', alpha=0.33, linestyle='-' , lw=4)
plt.grid(b=True, which="minor", color='gray', alpha=0.22, linestyle='--', lw=1)
plt.axis('tight')
#plt.xlim((2,12))
#plt.ylim((-1, 1))

leg = plt.legend(loc='best', fancybox=True, fontsize=14)
leg.get_frame().set_alpha(0.75)

plt.subplot(212)
plt.plot(t[0:N], ndither[0:N], linestyle='-', lw=2, alpha=0.57, color='yellow',
             label='Dither')
plt.plot(t[0:N], n0[0:N], linestyle='-', lw=2, alpha=0.97, color='Green',
             label='Quant. Noise (no dither)')
plt.plot(t[0:N], n1[0:N], linestyle='-', lw=2, alpha=0.57, color='Red',
             label='Quant. Noise (w/ dither)')


plt.xlabel(r'Time [s]', fontsize=18)
#plt.ylabel(r'Strain', fontsize=18)

#plt.title("IR Transmission during ALS Scan");
plt.grid(b=True, which="major", color='gray', alpha=0.33, linestyle='-' , lw=4)
plt.grid(b=True, which="minor", color='gray', alpha=0.22, linestyle='--', lw=1)
plt.axis('tight')
#plt.xlim((2,12))
#plt.ylim((-1, 1))

leg = plt.legend(loc='best', fancybox=True, fontsize=14)
leg.get_frame().set_alpha(0.5)
#plt.savefig("TRY.pdf", bbox_inches='tight')

plt.show()
```

```python
nfft = 16 * fs


f, Pyy = sig.welch(y0, fs, nperseg=nfft)
f, Pqq = sig.welch(q0, fs, nperseg=nfft)
f, Pnn = sig.welch(n0, fs, nperseg=nfft)
f, Pdd = sig.welch(n1, fs, nperseg=nfft)


fig = plt.figure(figsize=(16, 9))

plt.loglog(f, np.sqrt(Pyy), linestyle='-', lw=2, alpha=0.90, color='Black',
             label="Signal")
plt.loglog(f, np.sqrt(Pqq), linestyle='-', lw=4, alpha=0.35, color='Orange',
             label="Quantized Signal")
plt.loglog(f, np.sqrt(Pnn), linestyle='-', lw=2, alpha=0.75, color='Red',
             label="Quantization Noise (no dither)")
plt.loglog(f, np.sqrt(Pdd), linestyle='-', lw=2, alpha=0.75, color='Purple',
             label="Quantization Noise (w/ dither)")

plt.xlabel(r'Frequency [Hz]', fontsize=18)
#plt.ylabel(r'Displacment [m/$\sqrt{Hz}$]', fontsize=18)

#plt.title("IR Transmission during ALS Scan");
plt.grid(b=True, which="major", color='gray', alpha=0.33, linestyle='-', lw=4)
plt.grid(b=True, which="minor", color='gray', alpha=0.22, linestyle=':', lw=2)
#plt.axis('tight')
plt.xlim((1, 333))
plt.ylim((1e-3, 1e5))

leg = plt.legend(loc='best', fancybox=True, fontsize=14)
leg.get_frame().set_alpha(0.5)
plt.show()
```
